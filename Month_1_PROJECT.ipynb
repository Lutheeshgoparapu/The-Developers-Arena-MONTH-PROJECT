```python
## WEEK 1 — PYTHON BASICS SOLUTIONS
#1) Temperature Converter
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print("Fahrenheit:", fahrenheit)
```

```python
#2) Simple Calculator
a = float(input("Enter number 1: "))
b = float(input("Enter number 2: "))
op = input("Enter operator (+,-,*,/): ")

if op == '+':
    print(a + b)
elif op == '-':
    print(a - b)
elif op == '*':
    print(a * b)
elif op == '/':
    print(a / b)
else:
    print("Invalid operator")
```

```python
#3) Client Project — Average Temperature Script
temps = [30, 32, 29, 31, 33]
avg = sum(temps) / len(temps)
print("Average temperature:", avg)
```

```python
# WEEK 2 — DATA STRUCTURES & FUNCTIONS SOLUTIONS
#1) Remove duplicates from a list
lst = [1, 2, 2, 3, 4, 4, 5]
unique = list(set(lst))
print(unique)
```

```python
#2) Count occurrences using dictionary
text = "banana"
count = {}

for ch in text:
    count[ch] = count.get(ch, 0) + 1

print(count)
```

```python
#3) Function + List Comprehension
def square_list(nums):
    return [x*x for x in nums]

print(square_list([1,2,3,4,5]))
```

```python
## WEEK 3 — NUMPY & PANDAS SOLUTIONS
# Importing
import numpy as np
import pandas as pd
```

```python
#1) NumPy – Basic Operations
arr = np.array([10, 20, 30, 40])
print(arr + 5)
print(arr * 2)
print(arr.mean())


```

```python
#2) Pandas – Create DataFrame
data = {
    "Name": ["A", "B", "C"],
    "Marks": [75, 85, 90]
}
df = pd.DataFrame(data)
print(df)

```

```python
#3) Handle Missing Values
df = pd.DataFrame({
    "Name": ["Sam", "John", "Mike"],
    "Age": [25, None, 30],
})

df["Age"] = df["Age"].fillna(df["Age"].mean())
print(df)

```

```python
#4) Groupby & Aggregation
df = pd.DataFrame({
    "Dept": ["IT", "IT", "HR", "HR"],
    "Salary": [50000, 60000, 45000, 55000]
})

result = df.groupby("Dept")["Salary"].mean()
print(result)
```

```python

## Client Project — Clean & Analyze Dataset
df = pd.DataFrame({
    "Product": ["A","B","C","A","C"],
    "Sales": [100, 150, None, 120, 200],
    "Region": ["North","South","East","North","East"]
})

df["Sales"] = df["Sales"].fillna(df["Sales"].mean())

summary = df.groupby("Product")["Sales"].sum()

print(df)
print("\nTotal sales per product:")
print(summary)
```

```python

# WEEK 4 — VISUALIZATION SOLUTIONS
## Imports
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1) Line Plot
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Line Plot")
plt.show()

```

```python
# 2) Histogram
data = [10,20,30,40,10,30,20,50]

plt.hist(data, bins=5)
plt.title("Histogram")
plt.show()

```

```python
# 3) Scatter Plot
x = [1,2,3,4,5]
y = [5,4,6,5,7]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()

```
